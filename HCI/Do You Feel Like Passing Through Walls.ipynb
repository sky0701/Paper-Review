## Do You Feel Like Passing Through Walls?: Effect of Self-Avatar Appearance on Facilitating Realistic Behavior in Virtual Environments

- Sence of Presence에 대한 논문

### Purpose
- In Virtual Reality, we can easily pass-through objects, especially walls. This can disrupt the immersive experience. In this paper, they assume that a greater sense of presence would discourage users from walking through walls, and conducted a experiment that controls the anthropomorphism and visibility of self-avatars.


### Experiment
- There are four groups differ by self-avatars below. Each participant must enter Room1~4 and press all numbered buttons in order. Mission can be carried out quickly by penetrating the walls. They focus on 4 parts below.


### Measurement
- The number of people who passed through the wall in each room.
- The time delay until the first penetration.
- Sobo (Sense of body ownership) and Presence.
- SCR analysis in room 4 (If you get closer to the wall in Room 4, you can see a graphic that looks like electricity flowing.)


### Results
- As expected, the full-body human avatar group had fewer passes through the wall and took a long time to pass through the wall. Also, SCR was the highest in full-body avatar. The person who is in the lower presence took less time to finish the mission. In Subjective ownership, it is higher in Controller and Robot group, which needs further research on the correlation between reality and ownership.


### Analysis
- In this paper, the 5 Measuring Method of Sense of Presence was well revealed, so it was a good paper to study.
1.	Psychological and subjective measures: Gonzalez-Franco and Peck’s avatar embodiment questionnaire and Slater-Usoh-Steed questionnaire for subjective evaluation of Sobo and Presence.
2.	Psychophysical measures: Cognitive load on passing walls or pushing threating buttons can be measured by questionnaire about “response to threat”.
3.	Physiological measure: SCR (Skin conductance response)
4.	Performance measures: Each group’s performance was measured by the number of people passing through the wall and the time until the first penetration.
5.	Reflex response: In Room 4, the reactive behavior to electricity graphics was measured by SCR. In addition, all designed experiments deal with how users unconsciously react to their self- avatar.


### Criticize
1.	Effectiveness: Experimentally demonstrate that the full-body avatar (realistic and visible) can give more immersive experiences from fewer passes through walls.
2.	Limitations: More realistic graphics and improved avatar sync can increase confidence of experiment. Also, as four Rooms are tested sequentially, participants can get used to passing through the wall.
3.	Validity: In the case of full body avatar, the object has straight palms. It might give participant to have sense of difference, who was grabbing the controller. I wonder whether the questionnaire on presence or ownership is still valid.
